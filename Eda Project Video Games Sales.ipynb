Exploratory Data Analysis is a process to perform initial investigations on a data set so as to discover patterns, spot anomalies or validate assumptions.

In this process, an analyst relies on summary statistics and visualizations.

Important skills for conducting EDA with python include knowledge of python scripting, pandas and visualizations using matplotlib and/or seaborn

```python
#import the libaries to work with
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Data Set
The data set for this project is the "Video Game Sales Project" obtained from Kaggle.com. The dataset contains a list of video games with sales greater than 100,000 copies, generated by a scrape of vgchartz.com.

The data includes various attributes such as the name of the video game, platform, genre, sales in different regions (NA_Sales, EU_Sales, JP_Sales, Other_Sales, and Global_Sales), release year, and more.

This dataset provides a comprehensive understanding of the video game industry and will be used to analyze and visualize various trends and patterns in the sales of video games.

```python
#load the data set
data = pd.read_csv("vgsales.csv")
data.head()
```

## Cleaning and Manipulation
![image-3.png](data-cleaning1-1.jpg)

```python
#check for missing data using isna().sum()
data.isna().sum()
```

```python
#Since we know that the Year column has some missing data let's take a look it and correct that.
data[data.Year.isna()]
```

```python
#let's take out the rows with empty years
data_clean_Yr = data[data.Year.notna()]
data_clean_Yr
```

```python
#Let's do same for the Empty Publisher Values
data_clean_Yr_Pr = data[data.Publisher.notna()]
data_clean_Yr_Pr
```

```python
#Let's Give Our Columns More Befiting Names 
data_clean_Yr_Pr.rename(columns={'NA_Sales': 'North_America_Sales(m)', 'EU_Sales': 'Europe_Sales(m)', 'JP_Sales': 'Japan_Sales(m)', 
                                 'Other_Sales': 'Other_Sales(m)', 'Global_Sales': 'Total_Global_Sales(m)'}, inplace=True)
```

```python
#Let's make sure there are no missing values
data_clean_Yr_Pr.isna().sum()
```

```python
#Now let's take a look at what we've done
data_clean_Yr_Pr
```

```python
#First, let's create a pairplot for all the variables in our data set.
# Create a pair plot
sns.pairplot(data_clean_Yr_Pr)

# Show the plot
plt.show()
```

```python
#To get a closer look on corelations let's create an heatmap.
# Compute the correlation matrix
corr = data_clean_Yr_Pr.corr()

# Generate a heatmap of the correlation matrix
sns.heatmap(corr, annot=True)

# Show the plot
plt.show()
```

```python
#From the heatmaps we'll see that the ranks corelates with the sales, let's explore this using a bar chart.
Rank_sales = data_clean_Yr_Pr.groupby('Rank')['Total_Global_Sales(m)'].sum()

plt.bar(Rank_sales.index, Rank_sales.values)

# Rotate the x-axis labels
plt.xticks(rotation=90)

plt.ylim(0, 12.5)
plt.xlim(0, 1500)
# Add labels to the x and y axes
plt.xlabel('Rank')
plt.ylabel('Sales')

# Show the plot
plt.show()
```

```python
# We would notice the climb in sales between the year 2000-2010, Let's take a look at the publishers that released that year.
#Filter the DataFrame to only include the desired period of years using boolean indexing
data_period = data_clean_Yr_Pr[(data_clean_Yr_Pr.Year >= 2000) & (data_clean_Yr_Pr.Year <= 2010)]

#Group the filtered DataFrame by Publisher and calculate the total sales for each publisher
top_ten_publishers = data_clean_Yr_Pr.groupby('Publisher')['Total_Global_Sales(m)'].sum().nlargest(10)

#Sort the resulting Series in descending order to get the top performing publisher
top_publisher = top_ten_publishers.sort_values(ascending=False)
top_publisher

#Let's place the platforms in front of the top perfoming publishers to help make informed decisions.
for index, row in top_publisher.iteritems():
    publisher = index
    platform = data_clean_Yr_Pr[data_clean_Yr_Pr['Publisher'] == publisher]['Platform'].unique()[0]
    print(f"{publisher} ({platform})")
```

```python
# Now let's look at the diminishing sales between 2010-2020, Let's take a look at the the top performing publishers in terms of sales this year.
#Filter the DataFrame to only include the desired period of years using boolean indexing
data_period = data_clean_Yr_Pr[(data_clean_Yr_Pr.Year >= 2010) & (data_clean_Yr_Pr.Year <= 2020)]

#Group the filtered DataFrame by Publisher and calculate the total sales for each publisher
top_ten_publishers = data_clean_Yr_Pr.groupby('Publisher')['Total_Global_Sales(m)'].sum().nlargest(10)

#Sort the resulting Series in descending order to get the top performing publisher
top_publisher = top_ten_publishers.sort_values(ascending=False)
top_publisher

#Let's place the platforms in front of the top perfoming publishers to help make informed decisions.
for index, row in top_publisher.iteritems():
    publisher = index
    platform = data_clean_Yr_Pr[data_clean_Yr_Pr['Publisher'] == publisher]['Platform'].unique()[0]
    print(f"{publisher} ({platform})")
```

```python
# Let's take a look at all of the publishers and their genre.
for index, row in data_clean_Yr_Pr.groupby('Publisher').size().iteritems():
    publisher = index
    genres = data_clean_Yr_Pr[data_clean_Yr_Pr['Publisher'] == publisher]['Genre'].unique()
    print(f"{publisher} ({', '.join(genres)})")
```

```python
#Now we are trying get the year the top performing publishers released their games
# List of top publishers and their corresponding platforms
publishers_platforms = {"Nintendo": "Wii",
    "Electronic Arts": "PS4",
    "Activision": "X360",
    "Sony Computer Entertainment": "PS2",
    "Ubisoft": "Wii",
    "Take-Two Interactive": "PS3",
    "THQ": "PS",
    "Konami Digital Entertainment": "PS2",
    "Sega": "Wii",
    "Namco Bandai Games": "GBA"}

# Loop through each publisher-platform pair and print the release year for their games
for publisher, platform in publishers_platforms.items():
    # Select only the games from the publisher and platform of interest
    games = data_clean_Yr_Pr[(data_clean_Yr_Pr["Publisher"] == publisher) & (data_clean_Yr_Pr["Platform"] == platform)]
    
    # Print the release year for each game
    for index, row in games.iterrows():
        print(f"{publisher} ({platform}): {row['Year']}")
```
